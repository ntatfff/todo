```python
import plotly.graph_objects as go
import pandas as pd

df = pd.read_csv('4-tumor-bounding-box-2.csv')
df = df.loc[~df[['length','width', 'height']].apply(sorted, axis=1).duplicated()]

normal = df.where(df['color'] == 'normal')
x1 = normal['length']
y1 = normal['width']
z1 = normal['height']

mean = df.where(df['color'] == 'mean')
x2 = mean['length']
y2 = mean['width']
z2 = mean['height']

max = df.where(df['color'] == 'max')
x3 = max['length']
y3 = max['width']
z3 = max['height']

median = df.where(df['color'] == 'median')
x4 = median['length']
y4 = median['width']
z4 = median['height']

x5 = [140]
y5 = [140]
z5 = [140]

fig = go.Figure()
fig.add_trace(go.Scatter3d(x=x1, y=y1, z=z1, mode='markers', name='Normal', opacity=0.3, marker=dict(color='grey')))
fig.add_trace(go.Scatter3d(x=x2, y=y2, z=z2, mode='markers', name='Mean', marker=dict(color='orange')))
fig.add_trace(go.Scatter3d(x=x3, y=y3, z=z3, mode='markers', name='Max', marker=dict(color='red')))
fig.add_trace(go.Scatter3d(x=x4, y=y4, z=z4, mode='markers', name='Median', marker=dict(color='blue')))
fig.add_trace(go.Scatter3d(x=x5, y=y5, z=z5, mode='markers', name=' ', opacity=0.0001))

camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    # eye=dict(x=1.25, y=1.25, z=1.25)
    # eye=dict(x=1.25, y=-1.25, z=1.25)
    # eye=dict(x=-1.25, y=-1.25, z=1.25)
    # eye=dict(x=1.25, y=1.25, z=-1.25)
    # eye=dict(x=1.25, y=-1.25, z=-1.25)
    eye=dict(x=-1.25, y=-1.25, z=-1.25)
)

fig.update_layout(
    autosize=False,
    width=1000,
    height=1000,
    title='',
    scene=dict(
        xaxis_title='Length',
        yaxis_title='Width',
        zaxis_title='Height'
    ),
    scene_camera=camera
)
fig.update_traces(marker=dict(size=5))

fig.show()
```

```python
import plotly.graph_objects as go

# Sample data
x1 = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
z1 = [5, 10, 15, 20, 30]
y2 = [50, 40, 30, 20, 10]
z2 = [25, 20, 15, 10, 5]

# Create figure
fig = go.Figure()

# Add 3D scatter traces
fig.add_trace(go.Scatter3d(x=x1, y=y1, z=z1, mode='markers', name='Trace 1'))
fig.add_trace(go.Scatter3d(x=x1, y=y2, z=z2, mode='markers', name='Trace 1'))

# Set layout with uirevision
fig.update_layout(
    uirevision='constant',
    title='Prevent Axis Rescale in 3D Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show figure
fig.show()
```
