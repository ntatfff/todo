```python
import plotly.graph_objects as go
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/ntatfff/todo/refs/heads/202411241258/4-tumor-bounding-box-2.csv')
df = df.loc[~df[['length','width', 'height']].apply(sorted, axis=1).duplicated()]

normal = df.where(df['color'] == 'normal')
x1 = normal['length']
y1 = normal['width']
z1 = normal['height']

mean = df.where(df['color'] == 'mean')
x2 = mean['length']
y2 = mean['width']
z2 = mean['height']

max = df.where(df['color'] == 'max')
x3 = max['length']
y3 = max['width']
z3 = max['height']

median = df.where(df['color'] == 'median')
x4 = median['length']
y4 = median['width']
z4 = median['height']

fig = go.Figure()
fig.add_trace(go.Scatter3d(x=x1, y=y1, z=z1, mode='markers', name='Normal', opacity=0.3, marker=dict(color='grey')))
fig.add_trace(go.Scatter3d(x=x2, y=y2, z=z2, mode='markers', name='Mean(43, 43, 43)', marker=dict(color='orange')))
fig.add_trace(go.Scatter3d(x=x3, y=y3, z=z3, mode='markers', name='Max(132, 132, 132)', marker=dict(color='red')))
fig.add_trace(go.Scatter3d(x=x4, y=y4, z=z4, mode='markers', name='Median(42, 42, 42)', marker=dict(color='blue')))

# to prevent the plot scale when hide markers
x5 = [0, 140]
y5 = [0, 140]
z5 = [0, 140]
fig.add_trace(go.Scatter3d(x=x5, y=y5, z=z5, mode='markers', name=' ', opacity=0.0001))

camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    # eye=dict(x=1.5, y=1.5, z=1.5)
    eye=dict(x=1.5, y=-1.5, z=1.5)
    # eye=dict(x=-1.5, y=-1.5, z=1.5)
    # eye=dict(x=1.5, y=1.5, z=-1.5)
    # eye=dict(x=1.5, y=-1.5, z=-1.5)
    # eye=dict(x=-1.5, y=-1.5, z=-1.5)
)


fig.update_layout(
    autosize=False,
    width=1000,
    height=1000,
    title='',
    scene=dict(
        xaxis_title='Length',
        yaxis_title='Width',
        zaxis_title='Height',
    ),
    scene_camera=camera,
    font=dict(
        size=16
    ),
)
fig.update_traces(marker=dict(size=6))

fig.show()
```
